As you can see above the date column is not a date datatype, I will need to change that so that data stays in order.

```python
#using pandas built in to_datetime function to change the dateRep column form object to a datetime object
df['dateRep'] = pd.to_datetime(df['dateRep'], format='%d/%m/%Y')
df
```

```python
# we will drop the columns we do not need
reduced_columns = df.drop(['day', 'month', 'year','geoId', 'countryterritoryCode', 'continentExp'], axis=1)
```

```python
# head of our new slimed down dataframe
reduced_columns.head()
```

```python
# final 5 columns
reduced_columns.tail()
```

```python
# we will now drop the deaths column
reduced_columns_cases = reduced_columns.drop(['deaths'], axis=1)
reduced_columns_cases.head()
```

```python
reduced_columns_cases['casesPer100KPop'] = reduced_columns_cases.cases * 100000 / reduced_columns_cases.popData2019
reduced_columns_cases
```

```python
reduced_columns_cases.set_index('dateRep')
```

```python
cases = reduced_columns_cases.pivot(index='dateRep', columns='countriesAndTerritories', values='casesPer100KPop')
cases.head()
```

```python
cases.fillna(0, inplace=True)
cases
```

```python
cases_14_days = cases.iloc[-28:,:]
cases_14_days
```

```python
# 
cases_14_days_rolling_average = cases_14_days.rolling(window=14).mean().round(decimals=2)
cases_14_days_rolling_average
```
